```python
from pybaseball import statcast,statcast_batter, playerid_lookup
import pandas as pd
import numpy as np
```

```python
vaughn_info = pyb.playerid_lookup('vaughn', 'andrew')
print(vaughn_info)
```

```python
vaughn_id=683734
sox_start='2025-03-27'
sox_end='2025-05-21'
brew_start='2025-07-07'
brew_end='2025-10-17'
statcast_sox=statcast_batter(start_dt=sox_start,end_dt=sox_end,player_id=vaughn_id)
statcast_brew=statcast_batter(start_dt=brew_start,end_dt=brew_end,player_id=vaughn_id)
```

```python
statcast_sox['hc_y'].head()
statcast_sox['adj_x'] = statcast_sox['hc_x'] - 128.4
statcast_sox['adj_y'] = 208.2 - statcast_sox['hc_y']
statcast_sox['spray_angle'] = np.arctan2(statcast_sox['adj_x'], statcast_sox['adj_y']) * (180 / np.pi)
statcast_brew['hc_y'].head()
statcast_brew['adj_x'] = statcast_brew['hc_x'] - 128.4
statcast_brew['adj_y'] = 208.2 - statcast_brew['hc_y']
statcast_brew['spray_angle'] = np.arctan2(statcast_brew['adj_x'], statcast_brew['adj_y']) * (180 / np.pi)
```

```python
oppo=statcast_sox[(statcast_sox['spray_angle']>=15) & (statcast_sox['spray_angle']<=45)]
cen=statcast_sox[(statcast_sox['spray_angle']<15) & (statcast_sox['spray_angle']>=-15)]
pull=statcast_sox[(statcast_sox['spray_angle']<-15) & (statcast_sox['spray_angle']>=-45)]
oppo_brew=statcast_brew[(statcast_brew['spray_angle']>=15) & (statcast_brew['spray_angle']<=45)]
cen_brew=statcast_brew[(statcast_brew['spray_angle']<15) & (statcast_brew['spray_angle']>=-15)]
pull_brew=statcast_brew[(statcast_brew['spray_angle']<-15) & (statcast_brew['spray_angle']>=-45)]
```

```python
sox_oppo_woba=oppo['estimated_woba_using_speedangle'].mean()
sox_cen_woba=cen['estimated_woba_using_speedangle'].mean()
sox_pull_woba=pull['estimated_woba_using_speedangle'].mean()
oppo_ev80 = np.percentile(oppo['launch_speed'], 80)
cen_ev80 = np.percentile(cen['launch_speed'], 80)
pull_ev80 = np.percentile(pull['launch_speed'], 80)
sox_oppo_actual_woba=oppo['woba_value'].mean()
brew_oppo_woba=oppo_brew['estimated_woba_using_speedangle'].mean()
brew_cen_woba=cen_brew['estimated_woba_using_speedangle'].mean()
brew_pull_woba=pull_brew['estimated_woba_using_speedangle'].mean()
oppo_ev80_brew = np.percentile(oppo_brew['launch_speed'], 80)
cen_ev80_brew = np.percentile(cen_brew['launch_speed'], 80)
pull_ev80_brew = np.percentile(pull_brew['launch_speed'], 80)
print("White Sox")
print(sox_oppo_woba)
print(sox_oppo_actual_woba)
print(oppo_ev80)
print(sox_cen_woba)
print(cen_ev80)
print(sox_pull_woba)
print(pull_ev80)
print("Brewers")
print(brew_oppo_woba)
print(oppo_ev80_brew)
print(brew_cen_woba)
print(cen_ev80_brew)
print(brew_pull_woba)
print(pull_ev80_brew)
```
